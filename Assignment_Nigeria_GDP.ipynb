```python
import pandas as pd

df = pd.read_csv('Nigeria GDP.csv', skiprows=4)
```

```python
# checking for missing values

df.isnull().sum()
```

```python
# fixing missing value from the dataset

df = df.fillna(df.mean(numeric_only=True))
```

```python
# checking data type

df.dtypes
```

```python
df['1960'] = pd.to_numeric(df['1960'], errors='coerce')
```

```python
# Checking for duplicates

df.duplicated().sum()
```

```python
# Data Visualization

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.iloc[0]
```

```python
# Convert years into a usable format

country_data = df.iloc[0, 4:]  # adjust if needed
country_data = country_data.dropna()
country_data.index = country_data.index.astype(int)
```

```python
# Histogram

plt.hist(country_data.values, bins=10)
plt.title("GDP Distribution")
plt.show()
```

```python
# Scatter plot

plt.scatter(country_data.index, country_data.values)
plt.title("GDP Over Time (Scatter)")
plt.xlabel("Year")
plt.ylabel("Value")
plt.show()
```

```python
# Bar Chart

plt.bar(country_data.index, country_data.values)
plt.title("GDP by Year (Bar Chart)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Box Plot

plt.boxplot(country_data.values)
plt.title("GDP Distribution (Box Plot)")
plt.show()
```

```python
# Correlation Heat map

plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
